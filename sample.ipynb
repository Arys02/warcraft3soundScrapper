```python
from urllib.parse import urljoin, unquote
import os
import requests
from bs4 import BeautifulSoup
```

```python
def download_wowwiki_ogg(url):
    reponse_page = requests.get(url)
    soup_page = BeautifulSoup(reponse_page.text, 'html.parser')
    titre_page = soup_page.find('h1').text
    nom_dossier = titre_page.replace('/', '-').replace(':', '-').split('\n')[1]
    os.makedirs(nom_dossier, exist_ok=True)

    audios = soup_page.find_all('audio', src=True)
    for audio in audios:
        src = audio['src']
        
        src_complet = urljoin(url, src)
        src_complet = unquote(src_complet) 
        print(src_complet)# Décodage de l'URL
        reponse_audio = requests.get(src_complet, stream=True)
        nom_fichier = src_complet.split('/')[7]
        print(nom_fichier) 
        print(nom_dossier)
        chemin_fichier = os.path.join(nom_dossier, nom_fichier)

        with open(chemin_fichier, 'wb') as fichier:
            for chunk in reponse_audio.iter_content(chunk_size=128):
                fichier.write(chunk)
        print(f'Fichier téléchargé : {chemin_fichier}')
```

```python
urls = ['https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Humain',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Orc',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Elfe_de_la_nuit',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Mort-vivant',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Naga',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/D%C3%A9mon',
        'https://wowwiki.fandom.com/fr/wiki/R%C3%A9pliques_de_Warcraft_III/Neutre'
        ]
```

```python
for url in urls : 
    download_wowwiki_ogg(url)
```
